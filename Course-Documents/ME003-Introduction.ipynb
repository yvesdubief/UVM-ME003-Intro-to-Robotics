### Statement

The Syracuse sequence yields a simple yet unsolved mathematical problem: Given a starting integer $u_0\leq 1$, the sequence

$$
u_{n+1}=\begin{cases}\cfrac{u_n}{2}&\text{ if $u_n$ is even}\\
3u_n+1&\text{ if $u_n$ is odd}
\end{cases} \text{ for any $n\leq1$}
$$

is conjectured to always converged to the pattern (4,2,1) which repeats to infinity. This conjecture has not been formally proven and it is not our concern (we are engineers, not mathematicians). 

### Tasks

<ol>
    
<li>Write a python code that can study the Syracuse sequence for any random see (which has to be a positive integer)</li>
<li>Plot the number of iterations for integers between 1 and N (N>1) necessary for the sequence to reach 1</li>   
    
</ol>

```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
u = np.array([34])
print(u,u[-1])

while u[-1] != 1:
    if u[-1]%2 == 0:
        u = np.append(u,[u[-1]//2])
    else:
        u = np.append(u,[3*u[-1]+1])
plt.plot(u)
plt.show()
```

```python
N = 128
u0 = np.array([n+1 for n in range(N)])
```

```python
def syracuse(u_0):
    n = 0
    u = np.array([u_0])
    while u[-1] != 1:
        if u[-1]%2 == 0:
            u = np.append(u,[u[-1]//2])
        else:
            u = np.append(u,[3*u[-1]+1])
        n += 1
    return n
```

```python
nsteps = np.zeros(N)
for i in range(N):
    nsteps[i] = syracuse(u0[i])
    
plt.plot(u0,nsteps)
plt.show()
```
